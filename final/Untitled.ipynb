```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = 'C:/Users/Alex/Downloads/First_Results/First_Results/Low_LR'
name = 'training'
col = 'episode_moving_average_30'

dqn = pd.read_csv(path + f'/DQN/21000_{name}_metrics.csv')
dqn_ms = pd.read_csv(path + f'/DQN_MS/16000_{name}_metrics.csv')
ddqn = pd.read_csv(path + f'/DDQN/14000_{name}_metrics.csv')
ddqn_ms = pd.read_csv(path + f'/DDQN_MS/16000_{name}_metrics.csv')
ddqn_b = pd.read_csv(path + f'/_DDQN_Brian/15000_{name}_metrics.csv')
duel = pd.read_csv(path + f'/DuelingDQN/15000_{name}_metrics.csv')

plt.plot(dqn['episode'], dqn[col], label = 'DQN')
plt.plot(dqn_ms['episode'], dqn_ms[col], label = 'DQN Multi-step')
plt.plot(ddqn['episode'], ddqn[col], label = 'Double DQN')
plt.plot(ddqn_ms['episode'], ddqn_ms[col], label = 'Double DQN Multi-step')
plt.plot(ddqn_b['episode'], ddqn_b[col], label = 'Doulbe DQN ex Action Separation')
plt.plot(duel['episode'], duel[col], label = 'Dueling DQN')
plt.xlabel('Training Episodes')
plt.ylabel('Training Rewards')
plt.title('Low Learning Rate DQN Family on Breakout')
plt.legend(loc = 'best')
plt.show()
```
